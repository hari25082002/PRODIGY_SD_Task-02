```python
import tkinter as tk
import random as R
import tkinter.messagebox as mb
from tkinter import ttk

class GuessingGame():
    def __init__(self):
        self.attempts = 0
        self.max_attempts = None
        self.secret_number = None
        self.score = None

        self.T = tk.Tk()
        self.T.title("Number guessing game")
        
        self.create_widgets()
        self.enable_entry(None)
        
    def confirm_exit(self):
        response = mb.askyesno("Exit", "Are you sure you want to quit?")
        if response:
            self.T.destroy()

    def create_widgets(self):
        self.label = tk.Label(self.T, text="Guess a number between 1 and 100:")
        self.label.pack(padx=20, pady=20)

        self.F = tk.Frame(self.T)
        self.F.pack(padx=20)

        self.label_2 = tk.Label(self.F, text="Select the level", font=("Arial", 15))
        self.label_2.grid(row=0, column=0, padx=10, pady=5)

        # Define level constants for readability
        EASY, MEDIUM, HARD = 'Easy', 'Medium', 'Hard'

        levels = [EASY, MEDIUM, HARD]
        self.levels = ttk.Combobox(self.F, values=levels)
        self.levels.grid(row=0, column=1, padx=10, pady=5)
        
        
        self.label_3=tk.Label(self.F,text="",font=("Times New Roman",14))
        self.label_3.grid(row=1,columnspan=3,padx=10,pady=5)
        
        self.attempts_label = tk.Label(self.F, text="", font=("Times New Roman", 14))
        self.attempts_label.grid(row=2, columnspan=3, padx=10, pady=5)
        
        

        self.entry = tk.Entry(self.F, width=30 ,state="disabled")
        self.entry.grid(row=3, columnspan=3, padx=10, pady=5)

        self.button = tk.Button(self.F, text="Submit", command=self.check_guess)
        self.button.grid(row=4, columnspan=3, padx=10, pady=5)
        self.entry.bind('<Return>', lambda event=None: self.check_guess())

        
        self.levels.bind("<<ComboboxSelected>>", self.enable_entry)
        self.T.protocol("WM_DELETE_WINDOW", self.confirm_exit)
        
    def enable_entry(self, event):
            selected_level = self.levels.get()
            self.selected_level = selected_level
            if selected_level == 'Easy':
                self.label_3.config(text="Enter the number between (1-50)")
            elif self.selected_level == 'Medium':
                self.label_3.config(text="Enter the number between (51-70)")
            elif selected_level == 'Hard':
                self.label_3.config(text="Enter the number between (71-150)")
                
                
            if selected_level:
                self.entry.config(state="normal")
                self.levels.config(state="disabled")
            else:
                self.entry.config(state="disabled")
                self.entry.delete(0, tk.END)
        
        
            
       
            
            

    def check_guess(self):
        try:
            self.selected_level = self.levels.get()
            if self.selected_level == 'Easy':
                self.max_attempts = 10
                
                self.secret_number = R.randint(1, 50)
                
            elif self.selected_level == 'Medium':
                self.max_attempts = 5
                
                self.secret_number = R.randint(51, 70)
            else:
                self.max_attempts = 3
                
                self.secret_number = R.randint(71, 150)

            guess = int(self.entry.get())
            self.attempts += 1
            attempts_left = self.max_attempts - self.attempts
            self.attempts_label.config(text=f"Attempts left: {attempts_left}")

            if guess == self.secret_number:
                self.score = (self.attempts / self.max_attempts) * 100
                self.display_result("Congratulations! You guessed the number in {} attempts. Your score is {:.2f}%.".format(self.attempts, self.score))
                self.entry.delete(0,tk.END)
                self.restart_game()
            elif guess < self.secret_number:
                self.display_result("Too low. Try again.")
                self.entry.delete(0,tk.END)
            else:
                self.display_result("Too high. Try again.")
                self.entry.delete(0,tk.END)

            if self.attempts >= self.max_attempts:
                self.display_result("Sorry, you have reached the maximum number of attempts for {} level. The secret number was {}.".format(self.selected_level, self.secret_number))
                self.restart_game()
        except ValueError:
            self.display_result("Enter a valid number (1-100)")
            self.entry.delete(0,tk.END)

    def display_result(self, message):
        mb.showinfo("Result", message)

    def restart_game(self):
        response = mb.askyesno("Restart", "Do you want to play again?")
        if response:
            self.T.destroy()
            game = GuessingGame()
            game.T.mainloop()
        else:
            self.T.destroy()

if __name__ == "__main__":
    game = GuessingGame()
    game.T.mainloop()
```
